### Изменение концентрации металлов (Au, Ag, Pb) на различных этапах очистки. Описание выводов.

```python
steps = ['rougher', 'primary_cleaner', 'final']
types = ['output.concentrate']
metals = ['au', 'ag', 'pb']
```

```python
def concentration(step):
    data = pd.DataFrame()
    if step == 'rougher':
        for metal in metals:
            str_ = step + '.' + 'input.feed' + '_' + metal
            data[str_] = data_full.loc[:, str_]
    for t in types:
        for metal in metals:
            str_ = step + '.' + t + '_' + metal
            data[str_] = data_full.loc[:, str_]
    return data
```

```python
third_step = concentration(steps[2])
print(third_step.info())
third_step.hist(bins=500, figsize=(10, 10));
```

### Исследование и обработка аномалий по каждому веществу

```python
def search_anomalies(metals, step):
    if step == 'rougher':
        for i in metals:
            sns.displot(data_train[step + '.' + 'input.' + 'feed_' + i], kde=True, aspect=3);
            sns.displot(data_train[step + '.' + 'output.concentrate_' + i], kde=True, aspect=3);
    else:
        for i in metals:
            sns.displot(data_train[step + '.' + 'output.concentrate_' + i], kde=True, aspect=3);
```

```python
def describe_anomalies(data_train, metals, step):
    #     if step == 'rougher':
    for i in metals:
        print(f'Количество аномальных значений концентрации для этапа {step} ({i}): ' \
              f'{len(data_train[data_train[step + ".output.concentrate_" + i] == 0])}')
        print(f'Процент аномальных значений концентрации для этапа {step} ({i}): ' \
              f'{len(data_train[data_train[step + ".output.concentrate_" + i] == 0]) / len(data_train[step + ".output.concentrate_" + i]):.4%}')
```

```python
def delete_anomalies(data_train, metals, step):
    for i in metals:
        print('До удаления аномалий: ')
        describe_anomalies(data_train, metals, step)
        data_train = data_train[data_train[step + '.output.concentrate_' + i] != 0]
        print('\nПосле удаления аномалий:')
        describe_anomalies(data_train, metals, step)
```

### Функция для вычисления итоговой sMAPE.

```python
def smape(target, prediction):
    return np.sum(np.abs(target - prediction) / (np.abs(target) + np.abs(prediction)) * 2) * 100 / len(target)
```

```python
def final_smape(smape_0, smape_1):
    return 0.25 * smape_0 + 0.75 * smape_1
```

```python
smape_score = make_scorer(smape, greater_is_better=False)
model_forest = RandomForestRegressor(random_state=12345)
params = {'max_depth': range(1, 5), 'n_estimators': range(60, 100, 10)}
gs_forest = GridSearchCV(model_forest, params, scoring=smape_score, cv=5)
gs_forest.fit(features_0, target_0)
print(gs_forest.best_params_)
print(abs(gs_forest.best_score_))
```

### Дерево решений
